```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage, AIMessage
from pprint import pprint
```

```python
model = ChatGroq(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    api_key=os.getenv("GROQ_API_KEY")
)
```

```python
agent = create_agent(model=model)
```

```python
response = agent.invoke({
    "messages": [
        HumanMessage(content="What is the capital of Maharashtra?"),
        AIMessage(content="The capital of Maharashtra is Mumbai."),
        HumanMessage(content="Tell me more about Mumbai.")
    ]
})

pprint(response)
```

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="What is the capital of Maharashtra?")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True) 
```
